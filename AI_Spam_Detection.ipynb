```python
import pandas as pd
#df= pd.read_csv('EmailSpamDataset.csv', names=["Category","Message"])
df = pd.read_csv('SMSSpamCollection.txt', sep='\t', names=["label", "message"])
df.head()
```

```python
df.shape
```

```python
#Independant Features
#X=list(df['Message'])

X=list(df['message'])
X = X[1:]
```

```python
#Dependant Features
#y=list(df['Category'])
#y=list(df['label'])
```

```python
#Converting the Y into Weights
#y=list(pd.get_dummies(y,drop_first=True)['spam'])
y = list(df['label'].map({"ham": 0, "spam": 1}).dropna())
y = y[1:]
```

```python
# from sklearn.preprocessing import LabelEncoder

# label_encoder = LabelEncoder()
# y= label_encoder.fit_transform(y)
```

```python
#Train and Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state = 0)
```

```python
# intall the transformers to use the model
# !pip install transformers --upgrade
!pip install transformers==4.18.0
```

```python
from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
```

```python
y_test
```

```python
#Object Conversion
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))
```

```python
#from tftrainer import TrainArgument
#from tftrainer import TrainArgument, Trainer
#from transformers import TFDistilBertForSequenceClassification, TFTrainingArguments, Trainer
#import torch
#from transformers import TFDistilBertForSequenceClassification, BertForSequenceClassification
#from transformers import Trainer, TFTrainingArguments
#from transformers.trainer_tf import TFTrainer
#from transformers import TrainingArguments, Trainer
#from transformers import BertTokenizer, BertForSequenceClassification

from transformers import TFDistilBertForSequenceClassification, TFTrainer, TFTrainingArguments

training_args = TFTrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=2,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
)
```

```python
with training_args.strategy.scope():
    model = TFDistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

trainer = TFTrainer(
    model=model,                         # the instantiated ðŸ¤— Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset,            # evaluation dataset
)


```

```python
trainer.train()
```

```python
trainer.evaluate(test_dataset)
```
